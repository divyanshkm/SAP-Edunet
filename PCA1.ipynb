```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
df = load_breast_cancer(as_frame=True)
cancer = df.frame
cancer
```

```python
X = cancer[df['feature_names']]
X.shape
```

```python
X_mean = X.mean()
X_std = X.std()
Z= (X - X_mean)/X_std
Z
```

```python
c = Z.cov()
```

```python
sns.heatmap(c)
plt.show()
```

```python
eiganvalues, eiganvectors = np.linalg.eig(c)
print('Eigan values:\n', eiganvalues)
print('Eigan Values Shape:\n', eiganvalues.shape)
print('Eigan Vectors Shape:\n', eiganvectors.shape)
```

```python
idx = eiganvalues.argsort()[::-1]
eiganvalues = eiganvalues[idx]
eiganvectors = eiganvectors[:,idx]
```

```python
explained_var = np.cumsum(eiganvalues) / np.sum(eiganvalues)
explained_var
```
